```python
#import libraries

from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
#connect to website


URL = "https://www.amazon.com/Funny-Data-Accountant-Science-Business/dp/B0BLVP2TZL"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")
 
```

```python
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
```

```python
title = soup2.find("span", id="productTitle").get_text(strip=True)
```

```python
symbol = soup2.find("span", class_="a-price-symbol").get_text(strip=True)
whole = soup2.find("span", class_="a-price-whole").get_text(strip=True)
fraction = soup2.find("span", class_="a-price-fraction").get_text(strip=True)


price = f"{symbol}{whole}.{fraction}"
price = price.strip().replace("\n", "")

```

```python
print(title)
print(price)
```

```python
price = price.strip()[1:]
title = title.strip()
print(title)
print(price)
```

```python
def check_price():

    URL = "https://www.amazon.com/Funny-Data-Accountant-Science-Business/dp/B0BLVP2TZL"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")

    title_tag = soup.find("span", id="productTitle")
    price_tag = soup.find("span", class_="a-offscreen")

    # ðŸš¨ Stop if Amazon blocks you
    if not title_tag or not price_tag:
        print("Blocked by Amazon â€” retrying later")
        return

    title = title_tag.get_text(strip=True)
    price = price_tag.get_text(strip=True)[1:]  # remove $

    from datetime import date
    today = date.today()

    data = [title, price, today]

    with open('AmazonWebScraperDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

    print(title, price)

 
```

```python
while(True):
     check_price()
     time.sleep(86400)
```
